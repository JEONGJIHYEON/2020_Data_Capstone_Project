```python
# 필요한 패키지 불러오기
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
sales_df = pd.read_csv('C:/Users/jjh99/바탕 화면/capstone_data/real_data_no_log.csv', engine='python')
sales_df.shape # row와 column의 수 1301 80
```

```python
sales_df.info()
```

```python
# 상권 코드는 군집화에 아무런 영향을 미치지 않으므로 제거
sales = sales_df.drop(sales_df.columns[[0]], axis='columns')
```

```python
# K-MEANS Clustering 진행, K = 3
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, random_state=0).fit(sales)
print(kmeans.labels_)
```

```python
# 기존 데이터에 cluster number column 추가
sales_df['cluster'] = kmeans.labels_
sales_df.head()
```

```python
sales_df.groupby('cluster').size()
```

```python
# 새로운 csv 파일 'k=5means_2.csv' 로 저장
sales_df.to_csv('C:\\Users\\jjh99\\바탕 화면\\capstone_data\\real_data_cluster_5.csv', index = False, encoding = 'utf-8-sig')
```
